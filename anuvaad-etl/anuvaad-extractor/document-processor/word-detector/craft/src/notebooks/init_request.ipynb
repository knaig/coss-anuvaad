```python
def get_file(name):
    file = {
      "file": {
        "identifier": str(uuid.uuid4()),
        "name": name,
        "format": "PDF"
      },
      "config": {
        "OCR": {
          "option": "HIGH_ACCURACY",
          "language": "en"
        }
      }
    }

    return file
```

```python
pdf_names = [pdf.split('/')[-1]  for pdf in glob.glob(path + '/*pdf' )]
```

```python
files =[]
for pdf_name in sorted(pdf_names):
    files.append(get_file(pdf_name))
    

    
```

```python
request_in = {
  "inputs": None
    ,
    "jobID": "BM-15913540488115873",
    "state": "INITIATED",
    "status": "STARTED",
    "stepOrder": 0,
    "workflowCode": "abc",
    "taskID":"aaabbbba",
    "tool": "GVOCR",
    "message":"OCR",
    "metadata": {
                    "module": "WORKFLOW-MANAGER",
                    "receivedAt": 15993163946431696,
                    "sessionID": "4M1qOZj53tIZsCoLNzP0oP",
                    "userID": "d4e0b570-b72a-44e5-9110-5fdd54370a9d"
                }
}
```

```python
def get_batches(files, window_size, request = request_in):
    request_list = []
    for step in range( int(len(files) /window_size)+1):
        files_subset = files[step*window_size: (step + 1) *window_size]
        request = copy.deepcopy(request_in)
        request["inputs"] = files_subset
        if len(request["inputs"])!=0:
            request_list.append(request)
        
    return request_list
```

```python
request_list = get_batches(files=files,window_size=1,request=request_in)
```
